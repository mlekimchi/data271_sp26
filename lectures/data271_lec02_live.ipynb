```python
# Basic structure
class MyClass:
    def __init__(self, attribute1, attribute2):
        self.attribute1 = attribute1  # Instance attribute
        self.attribute2 = attribute2  # Instance attribute

    def describe(self):
        return self.attribute1, self.attribute2

instance = MyClass("value1", "value2")
instance.describe()
```

```python
# Example: use a class to represent a point in 2D space and define operations like distance
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_to(self, other):
        """Calculate the distance to another point."""
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2)**0.5

# Example usage:
p1 = Point(0, 0)
p2 = Point(3, 4)
p1.distance_to(p2)
```
