## Methods

```re.search(pattern, text)``` returns a match object

```re.findall(pattern, text)``` returns a list

```re.split(pattern_used_to_split, text)``` returns a list

```re.sub(old_pattern, new_pattern, text)``` returns a string

### ```re.search(pattern, text)```
Returns the index and not much else. It can be good to search large texts just to see if the pattern is there :/

````python
text = 'Tomorrow, row row your boat gently down the stream'

re.search('row',text)
````

````python
# example with no match
re.search('tsunami',text)

#print(re.search('tsunami',text))
````

### ```re.split(pattern_to_split_on, text)```
This splits your text based on the pattern you set. It returns a list.

````python
re.split(' ', text)
````

````python
re.split('row', text)
````

### ```re.sub(old_pattern, new_pattern, text)```

The output is a string with the substitutions as you would expect :)

````python
re.sub('Tomorrow', 'Today', text)
````

### ```re.findall(pattern, text)```

The most useful method. The method returns a list of all the matches. If you want to find how many matches, simply take the ```len(re.findall(pattern, text))```

````python
re.findall('row',text)
````

````python
# count the matches

len(re.findall('row',text))
````

#### ```re.findall('[]',text)```: using [] to search for a set of characters

````python
voice_message = 'Hi Emily, this is Ms Lark from the employment development department from the state of California. I was calling you regarding your current status and I wanted to find out if you have returned to work. If so, I need to know what employer you return to work to. Um, if you can give me a call back at 833-710-1801. Um, I can give you the next 24 hours, so tomorrow, bye.'
voice_message
````

````python
# without brackets
without_brackets = re.findall('abcz3', voice_message)
# with brackets
with_brackets = re.findall('[abcz3]', voice_message)

print(f'''without brackets: {without_brackets}
with brackets: {with_brackets}''')
````

Notice that if we do not use brackets, we are searching for the specific pattern of ```'abcz3'```. With the brackets, we are searching for the characters ```a, b, c, z, or 3```

#### Quesiton: how would you find all the numbers and what would the read out look like?

````python
re.findall('[0-9]', voice_message)
````

The returned list is not readable as a phone number so we will fix that later

#### Question: how to get the uppercase letters?

````python
re.findall('[A-Z]', voice_message)
````
