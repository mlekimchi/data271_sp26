## Methods

```re.search(pattern, text)``` returns a match object

```re.findall(pattern, text)``` returns a list

```re.split(pattern_used_to_split, text)``` returns a list

```re.sub(old_pattern, new_pattern, text)``` returns a string

### ```re.search(pattern, text)```
Returns the index and not much else. It can be good to search large texts just to see if the pattern is there :/

````python
text = 'Tomorrow, row row rowing your boat gently down the stream. I love owls'
voice_message = 'Hi Emily, this is Ms Lark from the employment development department from the state of California. I was calling you regarding your current status and I wanted to find out if you have returned to work. If so, I need to know what employer you return to work to. Um, if you can give me a call back at 833-710-1801. Um, I can give you the next 24 hours, so tomorrow, bye.'

re.search('row',text)
````

````python
# example with no match
re.search('tsunami',text)

#print(re.search('tsunami',text))
````

### ```re.split(pattern_to_split_on, text)```
This splits your text based on the pattern you set. It returns a list.

````python
re.split('\. ', voice_message)
````

````python
re.split('row', text)
````

### ```re.sub(old_pattern, new_pattern, text)```

The output is a string with the substitutions as you would expect :)

````python
re.sub('Tomorrow', 'Today', text)
````

### ```re.findall(pattern, text)```

The most useful method. The method returns a list of all the matches. If you want to find how many matches, simply take the ```len(re.findall(pattern, text))```

````python
re.findall('row',text)
````

````python
# count the matches

len(re.findall('row',text))
````

#### ```re.findall('[]',text)```: using [] to search for a set of characters

![image](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*Jna2tz8JyH6SH-nO_re99A.png)

````python
voice_message = 'Hi Emily, this is Ms Lark from the employment development department from the state of California. I was calling you regarding your current status and I wanted to find out if you have returned to work. If so, I need to know what employer you return to work to. Um, if you can give me a call back at 833-710-1801. Um, I can give you the next 24 hours, so tomorrow, bye.'
voice_message
````

````python
# without brackets
without_brackets = re.findall('abcz3', voice_message)
# with brackets
with_brackets = re.findall('[abcz3]', voice_message)

print(f'''without brackets: {without_brackets}
with brackets: {with_brackets}''')
````

Notice that if we do not use brackets, we are searching for the specific pattern of ```'abcz3'```. With the brackets, we are searching for the characters ```a, b, c, z, or 3```

#### Quesiton: how would you find all the numbers and what would the read out look like?

````python
re.findall('[0-9]', voice_message)
````

The returned list is not readable as a phone number so we will fix that later

#### Question: how to get the uppercase letters?

````python
re.findall('[A-Z]', voice_message)
````

## Meta Characters: set 1

```.``` - wildcard

```[ ]``` - set searching

```( )``` - return a group

```{ }``` - repetition

```^``` - first in string

```$``` - last in string

```*``` - zero or more

```?``` - optional

```+``` - one or more

### Meta character ```.``` : wildcard
Example, we want to pull out the periods from the voice message. As we saw, earlier, if we wanted to split the string by sentence (period), we couldn't just use a period, we had to use an excape character. That is because the period is a meta character

````python
# attempt #1 to pull out the periods in the voice message
re.findall('.', voice_message)
````

#### Question: What is happening?

The period is a meta character with the purpose of a wildcard. It will return *any and all* characters.

### 
Example using the meta character `.` as a wildcard.


````python
will = "Well well well if it isn't the consequences of my own whole actions, Mr Willy Wilhiem"

will1 = re.findall('w.ll',will)

will2 = re.findall('w..l',will)

will3 = re.findall('W...',will)


print(f''' 'w.ll': {will1}
'w..l': {will2}
'W...': {will3}
''')
````

### Meta character ```{}``` : repetition

The curly brackets indicate a repetition of whatever character or group that precedes.

````python
re.findall('W.{3}',will)
````

### Meta character ```|```: or

````python
book_review = "well myself and my counterpart on the dumb book club have finally finish this lovely book. i have so many feelings about it, but i wouldn't even know where to begin."

print(f'''Did the reviwer love or hate the book?

review text:
{book_review}''')
re.findall('love|hate',book_review)
````

### Meta character ```^``` : first

### Meta character ```$```: last

The carrot will return the pattern preceding it if it appears first in the text while the dolla sign will return the pattern if it is last in the string.


````python
# meta character ^pattern: first in string
#book_review

re.findall('^well', book_review)
````

````python
# meta character pattern$: last in string

re.findall('begin\.$', book_review)
````

### Matching repetitions of the preceeding pattern/character

### ```*``` - zero or more

### ```?``` - optional (zero or one match)

### ```+``` - one or more

#### Meta character ```*``` - zero or more

A greedy character!

````python
# * to match the pattern preceeding it!

# example, match a specific pattern: char*
#  'wo*' should match
#       w
#       wo
#       woo
#       wooooo....

tongue_twister = 'How much wood would a woodchuck chuck if a woodchuck could chuck wood? wooooohooo!'

re.findall('wo*',tongue_twister)
````

````python
# * to match the pattern preceeding it!

# example, match a wildcard: .*
#  'w.*' should match
#       w + any character that comes after!

tongue_twister = 'How much wood would a woodchuck chuck if a woodchuck could chuck wood? wooooohooo!'

re.findall('w.*',tongue_twister)
````

#### Meta character ```?``` - optional...zero or one repetition

````python
# * to match the pattern preceeding it!

# example, match a specific pattern: char*
#  'wo?' should match
#       w
#       wo
#       and that is all
# think of it as an optional 'o'

tongue_twister = 'How much wood would a woodchuck chuck if a woodchuck could chuck wood? wooooohooo!'

re.findall('wo?',tongue_twister)
````

````python
# example, a realer example

re.findall('https?', 'http https httpss')
````

## Meta Characters: Character Classes

![image](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*vlMw_nm9u9llM7X3qooz-w.png)

The uppercase version is the complement (opposite)!

### Character Class: ```\d``` : digits

````python
# example, pull out the numbers from the voice message!

voice_message
````

````python
re.findall('\d',voice_message)
````

#### Question: How would you get the phone number format of xxx-xxx-xxxx?

````python
re.findall('\d{3}-\d{3}-\d{4}',voice_message)
````

#### Question: How would you get just the area code?

````python
# use grouping ()

re.findall('(\d{3})-\d{3}-\d{4}',voice_message)
````

````python
# Example
#   grab all letters/numbers and zero or one times
#   the pattern ends when we hit a space!

re.findall('\w+',voice_message)
````

## In Class Practice

### Pull out just the numbers in the following text string:

```"There are 12 students, 3 TAs, and 1 instructor."```

#### Answer

````python
p1 = "There are 12 students, 3 TAs, and 1 instructor."

re.findall('\d', p1)
````

### Pull out the number *chunks* in the following text string:
```"There are 12 students, 3 TAs, and 1 instructor."```

#### Answer

````python
#"There are 12 students, 3 TAs, and 1 instructor."

# The + will get you a chunk of digits

re.findall('\d+', p1)
````

### Pull out the airport abbreviations

```"Flights today: SFO, LAX, JFK, and Alaska."```

#### Answer

````python
p2 = "Flights today: SFO, LAX, JFK, and sfo."

re.findall('[A-Z]{3}', p2)
````
