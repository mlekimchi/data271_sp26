## 2. Cleaning the Data

We'll work through four common cleaning steps: fixing column names, correcting data types, setting the index, and reordering.

### References

| Task | Code | Notes |
|------|------|-------|
| Rename specific columns | `df.rename(columns={'old':'new'})` | Returns new df unless `inplace=True` |
| Rename all columns | `df.columns = [...]` | Replaces all names at once |
| Drop column | `df.drop(columns='col', inplace=True)` | |
| Remove constant columns | `df.loc[:, df.nunique() > 1]` | Keep only columns with >1 unique value |
| Convert type | `df.astype({'col': int})` | |
| Parse datetime | `pd.to_datetime(df['col'])` | Use `format=` for non-standard strings |
| Set index | `df.set_index('col', inplace=True)` | |
| Sort rows | `df.sort_values(by='col', ascending=False)` | |
| Sort columns | `df.sort_index(axis=1)` | |

## 3. Reshaping Data

Data often arrives in the wrong shape for the analysis you want. Two transformations come up constantly: **wide → long** and **long → wide**. We'll build up each one step by step using weather data — then you'll practice on your own.

### References

| Direction | Function | When to use |
|-----------|----------|--------------|
| Wide → Long | `.melt(id_vars=, value_vars=)` | Each measurement → its own row |
| Wide → Long | `.stack()` | Column names become a new index level |
| Long → Wide | `.pivot(index=, columns=, values=)` | One value per cell — no aggregation |
| Long → Wide | `.pivot_table(aggfunc=)` | Multiple values per cell — aggregate them |

### Wide → Long

The weather dataset has one column per variable — let's collapse it so each row is a single measurement.

````python
df_weather_wide = pd.read_csv('sample_weather.csv')
df_weather_wide = df_weather_wide.iloc[:,1:]
df_weather_wide
````

````python
# METHOD 1

# transpose with more informative columns
date_index = df_weather_wide.set_index('date')
date_index
````

````python
# Make a long Series
#   the result is a multi-indexed Series
#   level 0 index   |   level 1 index   |   value
#   dates, set index|   column names    |   cell values


long_df = date_index.stack()
long_df
````

````python
# now we get rid of the double indexing
#    we reset the index

long_df = long_df.reset_index(name='value')
long_df

# and then we will rename that column from level_1 ---> variable
#long_df.rename(columns = {'level_1':'variable'},inplace=True)
````

````python
# Method 1.5: a shortcut 
#   we pre-name the second index lol and then we don't have to rename it later

date_index.columns.name = 'variable'
date_index
````

````python
# Now renaming isn't necessary
date_index.stack().reset_index(name = 'value')
````

### The Two methods, `df.stack()` and `df.melt()`

Both methods get you to the same place. Here's the full step-by-step for each:

---

**`df.stack()`** *(3–4 steps)*

| Step | Code | What it does |
|------|------|--------------|
| 1 | `df.set_index('date')` | Move the ID column into the index so only the *data* columns remain |
| 2 | `.stack()` | Collapse all remaining columns into rows — column names become a new index level |
| 3 | `.reset_index(name='value')` | Flatten the MultiIndex back to regular columns; `name=` sets the values column label |
| 4 | `.rename(columns={'level_1': 'variable'})` | Give the auto-generated `level_1` column a meaningful name |

> **Shortcut for step 4:** Set `df.columns.name = 'variable'` *before* stacking — then the level is already named and `.rename()` isn't needed.

---

**`df.melt()`** *(1 step)*

| Argument | Role | Example |
|----------|------|---------|
| `id_vars=` | Column(s) that stay as row identifiers — repeated for every melted variable | `'date'` |
| `value_vars=` | Columns to collapse into rows — their names become the `variable` column | `['max_temp', 'min_temp', 'inches_of_rain']` |

```python

date    max_temp  min_temp  inches_of_rain
↑           ↑        ↑            ↑
id_vars  ←————————— value_vars ————————→    # the values of each cell become a new column called 'value'
```


`df_weather_wide.melt(id_vars='date', value_vars=['max_temp', 'min_temp', 'inches_of_rain'])`

```python
date  |  variable       |  value
——————————————————————————————————
id_vars  |  col names      |  cell values
         |  from           |  from
         |  value_vars     |  value_vars

```

> **When to use which:**
> Use `.melt()` when you know your column names and want clean, readable code.
> Use `.stack()` when you want to collapse *all* columns at once, or when working with a MultiIndex.

````python
# METHOD 2

long_df = df_weather_wide.melt(id_vars = 'date',value_vars = ['max_temp','min_temp','inches_of_rain'])
long_df
````

### Long → Wide

Now let's go the other direction. `.pivot()` works great when there's exactly **one value per (index, column) pair** — but real data often has duplicates. Let's see what happens.

````python
# change long format back into wide format
long_df.pivot(index = 'date',columns = 'variable',values='value')
````

````python
long_df = pd.read_csv('long_data.csv')
long_df = long_df.iloc[:,1:]
long_df.head()
````

````python
# Pivot the data to get average sales by date and category
long_df.pivot_table(index=['date'], columns='category', values=['sales'])
````

````python
# Pivot the data to get TOTAL sales by date and category
wide_df = long_df.pivot_table(index=['date'], columns='category', values=['sales'], aggfunc=sum)
wide_df
````

````python
# Pivot the data to get TOTAL sales by date, product, and category
long_df.pivot_table(index='date', columns=['category','product'], values=['sales'], aggfunc=sum)
````

````python
# Go from wide to long
wide_df.reset_index().melt(id_vars='date', var_name=['type','category'])
````

## Appendix — Cleaning & Reshaping Quick Reference

### Cleaning
```python
df.rename(columns={'old':'new'}, inplace=True)       # rename specific columns
df.columns = ['col1', 'col2', ...]                   # rename all columns
df.drop(columns='col', inplace=True)                 # drop a column
df = df.loc[:, df.nunique() > 1]                     # drop constant columns
df = df.astype({'col': int})                         # convert type
df['col'] = pd.to_datetime(df['col'])                # parse datetime
df.set_index('col', inplace=True)                    # set index
df.sort_values(by='col', ascending=False, inplace=True)  # sort rows
```

### Reshaping
```python
# Wide → Long
df.melt(id_vars='id_col', value_vars=['a','b'])       # melt
df.set_index('id').stack().reset_index(name='value')  # stack

# Long → Wide
df.pivot(index='id', columns='var', values='val')     # pivot (no aggregation)
df.pivot_table(index='id', columns='var',             # pivot_table (aggregates)
               values='val', aggfunc=sum)
```
