## Matplotlib basics

Really all you need to get started is the **input** and the function to plot: ```plt.plot()``` and the function to show: ```plt.show()```

And `plt.figure(figsize=(lenght,width))` to change the dimensions of the figure

````python
# Whenever we want to use matplotlib
import matplotlib.pyplot as plt
````

````python
# If you have data living in lists
x = [1,2,3,4,5]
y = [1,4,7,3,5]

plt.plot(x,y) # generate the plot
plt.show() # display the plot
````

Example of multiple plots with the same ```x``` axis

````python
# Can also plot more than one line at a time
x = [1,2,3,4,5]
y = [1,4,7,3,5]
z = [1,5,7,2,7]
n = [7,3,2,5,2]

plt.plot(x,y) # generate the plot 
plt.plot(x,z) # generate the plot 
plt.show() # display the plot
````

Example of multiple plots but with different ```x``` axis. Notice the ```x``` window scales *sneakily* according to each different set of ```x```

````python
# if we have two different x-axes, the categorical get's plotted
x2 = ['a','b','c','d','e']
y2 = [1,6,2,5,1]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
````

## Plotting distributions 

### Bar Plot for category distributions: `plt.bar(categories, counts)`

Normally, DataFrames are records of occurances. When we make a bar plot, we are plotting the count of the occurance so we will have to probably use `df.col_name.value_counts()` to get the data in the form we desire

````python
# Import Affordability Data
df = pd.read_csv('https://raw.githubusercontent.com/bethanyj0/data271_sp24/main/midterms/property_price.csv')
df.head()
````

### Check in

How can we pull out just the state abreviation from the `City` column?

#### Answer

````python
[city[-2:] for city in df.City]

import re
[re.findall('[A-Z]{2}', city)[0] for city in df.City]
````

#### And now let's create a new column called `state`

````python
df['state'] = [city[-2:] for city in df.City]
df.head()
````

### Get counts: `df.col_name.value_counts()`
Use `df.state.value_counts()` to get the summary counts for each ocurance by `state`

## Object-Oriented Plotting

If you would like even more control over your figure, you can use object-oriented plotting.

Matplotlib graphics have 3 components:
- Figure: `fig` the entire space of the plot, like the blank canvas space
- Axes: `ax` the plotting region inside the figure
- plot: the data and elements inside the axes

Prior, we used `plt.plot()` silently creates default Figure and Axes objects for us.

Below is an example of how we can set our own parameters for even more control over our graphics:

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

````python
hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
scores = np.array([50, 55, 65, 70, 75, 85, 90, 95])
````

### Example of using default `Figure` and `Axes`

Matplotib automatically creates a figure and axes for us without having to specify.

````python
# Using the default

plt.plot(hours, scores)
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.title("Study Hours vs Test Score")
plt.show()
````

#### Same plot but with object-oriented plotting: `plt.subplots()`

We will use `plt.subplots()` which returns a `(Figure object, Axes object)`

We will manually create the `Figure` and the `Axes` object so we can modify it :)


#### Facets view with $(n \times m)$ subplots? `plt.subplots(n, m)`
If we want multiple plots in one figure, we can set the number of rows `n` and the number of columns `m` of subplots!

If we leave the function empty, we will just get one figure.

### Example with OOP, one plot
First, let's call `plt.subplots()`. You do not run it by itself but here I just want to show you what it does!

````python
# example wiith object oriented plotting

# first we set up the plotting space

# pull out the fig and ax so we can customize it!
fig, ax = plt.subplots()
````

#### Alternative to call `fig` and `ax` separately

````python
fig = plt.figure()
ax = fig.add_subplot(111)
````

####

````python
fig, ax = plt.subplots()
ax.plot(hours, scores)
ax.set_xlabel("Hours Studied")
ax.set_ylabel("Test Score")
ax.set_title("Study Hours vs Test Score")

plt.show()
````

### Example with OOP, multiply subplots!
The power now is that we can create subplots! We will create one row of plots with 2 side-by-side subplots:

`plt.subplots(1, 2)`

The Axes object that we pull out is an array:

`ax[0] =` axis information for the first subplot

`ax[1] =` axis information for the second subplot

````python
fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(hours, scores)
ax[0].set_title("Line Plot")

ax[1].scatter(hours, scores)
ax[1].set_title("Scatter Plot")

plt.show()
````

### Checkin: What if we wanted a $(2 \times 2)$ subplot?

How do we set up the plotting space with `plt.subplots()`?

### Checkin: Now, how can we add in our subplots?

I want first row to have linear, quadratic
and the second row to have square root, logarithm

*Hint*: you will refer to the `ax[row, col]`

#### Answer

````python
fig, ax = plt.subplots(2, 2, figsize=(8,6))

ax[0,0].plot('x', 'Linear', data=df)
ax[0,0].set_title("Linear")

ax[0,1].plot('x', 'Quadratic', data=df)
ax[0,1].set_title("Quadratic")

ax[1,0].plot('x', 'Square Root', data=df)
ax[1,0].set_title("Square Root")

ax[1,1].plot('x', 'Logarithm', data=df)
ax[1,1].set_title("Logarithm")

plt.tight_layout()
plt.show()
````

# Appendix

## Differences in `plt` versus `ax`

| Purpose              | `plt` (basic?)   | `ax` (object-oriented)     |
| -------------------- | --------------------- | -------------------------- |
| Create figure & axes | `plt.subplots()`      | `fig, ax = plt.subplots()` |
| Set figure size      | `plt.figure(figsize=(6,4))` | `fig, ax = plt.subplots(figsize=(6,4))` |
| Line plot            | `plt.plot(x, y)`      | `ax.plot(x, y)`            |
| Scatter plot         | `plt.scatter(x, y)`   | `ax.scatter(x, y)`         |
| Bar plot             | `plt.bar(x, y)`       | `ax.bar(x, y)`             |
| Histogram            | `plt.hist(data)`      | `ax.hist(data)`            |
| X-axis label         | `plt.xlabel("Label")` | `ax.set_xlabel("Label")`   |
| Y-axis label         | `plt.ylabel("Label")` | `ax.set_ylabel("Label")`   |
| Title                | `plt.title("Title")`  | `ax.set_title("Title")`    |
| Legend               | `plt.legend()`        | `ax.legend()`              |
| Set x-limits         | `plt.xlim(a, b)`      | `ax.set_xlim(a, b)`        |
| Set y-limits         | `plt.ylim(a, b)`      | `ax.set_ylim(a, b)`        |
| Grid                 | `plt.grid(True)`      | `ax.grid(True)`            |
| Show plot            | `plt.show()`          | `plt.show()`               |

### Plot subplots iteratively!!!
For a subplot with `r`=number of rows and `c`=number of columns for a total of $r\times c$ subplots.

We iterate through a list of the columns...minus the $x$ axis.

```python
fig, ax = plt.subplots(r, c, figsize=(8,6))

columns = list(df.columns.drop("x"))   # drop the independent variable

for i in range(r):
    for j in range(c):
        index = i*c + j
        ax[i,j].plot("x", columns[index], data=df)
        ax[i,j].set_title(columns[index])

plt.tight_layout()
plt.show()
```
