### Key tool: `re.findall()` with capturing groups
Parentheses `()` in a pattern create **capturing groups** â€” `re.findall()` returns just the captured parts.

```python
import re
# No groups: returns full matches
re.findall(r'\d+', 'I have 3 cats and 12 dogs')   # ['3', '12']

# One group: returns the captured part
re.findall(r'(\d+)', 'I have 3 cats and 12 dogs')  # ['3', '12']

# Multiple groups: returns list of tuples
re.findall(r'(\d+)-(\d+)', '12-05 and 03-22')     # [('12', '05'), ('03', '22')]
```

### The `axis=` parameter
- `axis=0` â†’ operate **down the rows** (result has one value per column)
- `axis=1` â†’ operate **across the columns** (result has one value per row)

```python
import numpy as np
a = np.array([[3, 7, 2],
              [5, 1, 8]])
a.max(axis=0)  # [5, 7, 8] â€” max of each column
a.max(axis=1)  # [7, 8]    â€” max of each row
a.sum(axis=1)  # [12, 14]  â€” sum of each row
```

### `.ravel()`, `.argsort()`, and fancy indexing
These three tools chain together to sort one array **by the order of another**:

```python
items  = np.array(['a', 'b', 'c'])
values = np.array([30, 10, 20])

# argsort() gives the indices that would sort the array
values.argsort()           # [1, 2, 0]  (10 is index 1, 20 is index 2, 30 is index 0)

# Use those indices to reorder items
items[values.argsort()]    # ['b', 'c', 'a']  â€” sorted by value ascending

# Flip with [::-1] for descending
items[values.argsort()[::-1]]  # ['a', 'c', 'b']  â€” sorted by value descending

# For 2D arrays, ravel() flattens first
items_2d = np.array([['x', 'y'], ['z', 'w']])
vals_2d  = np.array([[5, 1], [3, 9]])
items_2d.ravel()[vals_2d.ravel().argsort()[::-1]]  # ['w', 'x', 'z', 'y']
```

### `.argmax()` + fancy indexing
`.argmax(axis=1)` gives the **column index** of the max in each row. Combine with row indices to pull out those specific elements:

```python
prices = np.array([[6., 8., 10.],
                   [6., 7.,  7.],
                   [7., 6.,  5.]])

prices.argmax(axis=1)      # [2, 1, 0]  â€” column index of max per row

items = np.array([['A', 'B', 'C'],
                  ['D', 'E', 'F'],
                  ['G', 'H', 'I']])

# Fancy index: row [0,1,2], col [2,1,0]
items[[0,1,2], prices.argmax(axis=1)]  # ['C', 'E', 'G']
```
