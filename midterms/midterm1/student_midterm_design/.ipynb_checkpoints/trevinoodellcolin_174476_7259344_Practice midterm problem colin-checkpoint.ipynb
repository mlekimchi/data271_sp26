```python
snacks = {
    "popcorn": 8.50,
    "nachos": 7.25,
    "candy": 4.00,
    "soda": 5.75,
    "pretzel": 6.50,
    "water": 3.00,
    "hot dog": 6.25
}
```

a) Use dictionary indexing to find the price of nachos. Assign it to nacho_price.

```python
nacho_price = snacks["nachos"]
nacho_price
```

b) Use list comprehension to create a list called cheap_snacks containing the names of snacks that cost less than $6.00.

```python
cheap_snacks = [name for name, price in snacks.items() if price < 6.00]
cheap_snacks
```

c) Use a built-in Python function to compute the total cost if you bought every snack on the menu. Assign your result to total_cost.

```python
total_cost = sum(snacks.values())
total_cost
```
