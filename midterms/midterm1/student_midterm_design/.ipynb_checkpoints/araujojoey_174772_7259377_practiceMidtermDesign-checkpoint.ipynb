The question: Can you use regex to grab and change the formatting for dates? Dates have to end as MM-DD-YYYY. 

```python
import numpy as np
import re
from csv import reader
```

```python
water = open('water_data_dates.csv')
data = reader(water) # type is csv.reader
data = [i[0] for i in list(data)]
data[0]=re.sub("\ufeff","",data[0])
data
```

With dates above, make them all the same and sort afterwards

```python
# splitting dates with words and dates with numbers
words_match = r"(\w+)\s(\d+)\D*(\d{4})"
num_match = r"(\d+)[-/](\d+)[-/](\d+)"
dates_with_words = [i for i in data if re.findall(words_match,i)]
dates_with_num = [j for j in data if re.findall(num_match,j)]
#dates_with_words
#dates_with_num
```

```python
#making them all MM-DD-YYYY
months_m = {'January':'01','Jan':'01','Feb':'02','February':'02'}
formatted = []
#cleans dates with numbers 
for date in dates_with_num:
    g = re.search(num_match,date)
    m, d, y = g.groups()
    y = y if len(y) == 4 else '20' + y
    formatted.append(f"{m.zfill(2)}-{d.zfill(2)}-{y}")
#cleans dates with words 
for date in dates_with_words:
    g = re.search(words_match,date)
    m_name, d, y = g.groups()
    m_num = months_m[m_name]
    formatted.append(f"{m_num}-{d.zfill(2)}-{y}")
```
