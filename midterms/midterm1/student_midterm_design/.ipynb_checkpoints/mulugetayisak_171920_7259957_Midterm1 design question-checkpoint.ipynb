You are given the following NumPy array of student email addresses:

emails = np.array([
    "alice13@humboldt.edu",
    "bob_smith@gmail.com",
    "caro99@humboldt.edu",
    "david@yahoo.com",
    "eva24@humbolt.edu"
])

1.

A. Using regular expressions, filter the array to include only email addresses that:

End with "@humboldt.edu"

Store the result in a NumPy array called school_emails.

B. From the filtered results, extract only the username (the part before @) and store them in a NumPy array called usernames.

C.How many school email addresses are there?

```python
import numpy as np
import re

# Given array
emails = np.array([
    "alice13@humboldt.edu",
    "bob_smith@gmail.com",
    "caro99@humboldt.edu",
    "david@yahoo.com",
    "eva24@humbolt.edu"
])

# Part A: Filter emails ending with @school.edu
pattern = r"@humboldt\.edu$"

school_emails = np.array([e for e in emails if re.search(pattern, e)])

print("School Emails:")
print(school_emails)


# Part B: Extract usernames
usernames = np.array([e.split("@")[0] for e in school_emails])

print("\nUsernames:")
print(usernames)


# Part C: Count
count = len(school_emails)

print("\nTotal School Emails:")
print(count)
```
