Write one line of Python code using re.findall() to extract only the city names from address_list, in the same order as they appear.

```python
import re
```

```python
address_list = ["123 Main St Apt 4B, Arcata, CA 95521",
                "55-57 Broadway Suite 210, New York, NY 10012",
                "890 Pine Rd, Eureka, CA 95501",
                "742 Evergreen Terrace Unit 3, Springfield, OR 97477"]
```

```python
cities = re.findall(r",\s*([^,]+),\s*[A-Z]{2}\s*\d{5}", " ".join(address_list))
cities
```
