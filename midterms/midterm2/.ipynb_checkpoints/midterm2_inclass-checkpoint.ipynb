Run the two cells below, then for **each plot**: comment on the pros and cons of the visualization choice. What does it show well? What doesn't work, and what would you use instead?

```python
# Warm-up setup — run this cell first
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

wu_data = pd.DataFrame({
    'county': ['Alameda', 'Fresno', 'Humboldt', 'Los Angeles', 'Marin',
               'Sacramento', 'San Diego', 'Tulare'],
    'poverty_rate': [0.099, 0.212, 0.172, 0.147, 0.064, 0.134, 0.111, 0.228],
    'uninsured_rate': [0.047, 0.136, 0.103, 0.105, 0.033, 0.073, 0.088, 0.161]
})
```

```python
# Plot A — run this cell, do not modify
wu_data.plot.pie(y='poverty_rate', labels=wu_data['county'], figsize=(6, 6), legend=False)
plt.title('Poverty Rate by County (2022)')
plt.ylabel('')
plt.show()
```

```python
# Plot B — run this cell, do not modify
plt.figure(figsize=(10, 4))
plt.plot(wu_data['county'], wu_data['uninsured_rate'], marker='o')
plt.title('Uninsured Rate by County')
plt.show()
```
